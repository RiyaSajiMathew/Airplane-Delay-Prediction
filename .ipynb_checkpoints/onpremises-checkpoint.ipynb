#### To extract CSV files from the ZIP files

````python
def zip2csv(zipFile_name , file_path):
    
    """
    Extract csv from zip files
    
    zipFile_name: name of the zip file
    
    file_path : name of the folder to store csv
    
    """

    try:
        with ZipFile(zipFile_name, 'r') as z:
            
            print(f'Extracting {zipFile_name} ') 
                    
            z.extractall(path=file_path) 
            
    except:
        
        print(f'zip2csv failed for {zipFile_name}')
    
    

zip2csv(zip_path, base_path)

zip_files = os.listdir(csv_base_path)


for file in zip_files:
    
    zip2csv(csv_base_path + file, csv_base_path)

print("Files Extracted")


````

#### Combining all the CSV files


Steps:
First, create an empy dataframe that you will use to copy your individual dataframes from each file. 
Then, for each file in the `csv_files` list:

1. Read the CSV file into a dataframe  
2. Filter the columns based on the `filter_cols` variable

```
        columns = ['col1', 'col2']
        df_filter = df[columns]
```

3. Keep only the subset_vals in each of the subset_cols. Use the `isin` Pandas function ([documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isin.html)) to check if the `val` is in the dataframe column and then choose the rows that include it.

```
        df_eg[df_eg['col1'].isin('5')]
```

4. Concatenate the dataframe with the empty dataframe 

````python

def combine_csv(csv_files, filter_cols, subset_cols, subset_vals, file_name):
    
    """
    Combine csv files into one Data Frame
    csv_files: list of csv file paths
    filter_cols: list of columns to filter
    subset_cols: list of columns to subset rows
    subset_vals: list of values to subset rows
    """
    # Create an empty dataframe
    
    df = pd.DataFrame()
    
    # To read csv files into a dataframe 
    
    for files in csv_files:
        
        df = df.append(pd.read_csv(files))
        
        # Filtering dataframe columns
        
        df = df[filter_cols]
        
        # Subsetting the rows.We keep only subset_vals in each of the subset_cols.
        
        for cols, vals in zip(subset_cols, subset_vals):
            
            df = df[df[cols].isin(vals)]
            
        print(df)
            
        # saving the dataframe
        
        df.to_csv(file_name, index=False)
            
        
   
    
````

## Model evaluation
In this section, we'll evaluate our trained model on test data and report on the following metrics:
* Confusion Matrix plot
* Plot the ROC
* Report statistics such as Accuracy, Percision, Recall, Sensitivity and Specificity

To view a plot of the confusion matrix, and various scoring metrics, create a couple of functions:

````python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, target_predicted):
    
    cm = confusion_matrix(y_true, target_predicted)
    
    print('Confusion matrix\n\n', cm)
    
    print('\nTrue Positives(TP) = ', cm[0,0])
    
    print('\nTrue Negatives(TN) = ', cm[1,1])
    
    print('\nFalse Positives(FP) = ', cm[0,1])
    
    print('\nFalse Negatives(FN) = ', cm[1,0])
    
    # visualize confusion matrix with seaborn heatmap
    
    cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])
    
    sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
    
    print(classification_report(y_true, target_predicted))
    
````

````python
from sklearn import metrics

def plot_roc(test_labels, target_predicted):
    
    fpr, tpr, _ = metrics.roc_curve(test_labels,  target_predicted)
    
    auc = metrics.roc_auc_score(test_labels, target_predicted)
    
    #create ROC curve
    
    plt.plot(fpr,tpr,label="AUC="+str(auc))
    
    plt.ylabel('True Positive Rate')
    
    plt.xlabel('False Positive Rate')
    
    plt.legend(loc=4)
    
    plt.show()
````

To plot the confusion matrix, call the `plot_confusion_matrix` function on the `test_labels` and `target_predicted` data from our batch job:

To print statistics and plot an ROC curve, call the `plot_roc` function on the `test_labels` and `target_predicted` data from our batch job:

recall is 79% and is represented by the total number of true positives divided by the total number of actual positives.

We can see that accuracy and recall are almost same but precision of the model is very high. Precision shows how far off each individual â€œpredictionâ€ is. A high precision means that predictions are relatively close to each other. 


What can you summarize from the confusion matrix?


A confusion matrix is a summary of prediction results on a classification problem. It shows the ways in which our classification model gets confused when it makes predictions. It is said that a good model is one which has high TP and TN rates, while low FP and FN rates.
In our case, TP = 258234 but TN = 0. i.e, the model didn't predict the negative classes correctly. 

From our cinfusion matrix,

True Positives (TP) = 258234 (when the actual value is Positive and predicted is also Positive.)

True negatives (TN) = 0 (when the actual value is Negative and prediction is also Negative.)

False positives (FP) = 0  (When the actual is negative but prediction is Positive. Also known as the Type 1 error)

False negatives (FN) = 68884 (When the actual is Positive but the prediction is Negative. Also known as the Type 2 error)


The classifier made correct predictions for 258234 observations and made wrong predictions for 68884 observations.

Import weather data prepared for the airport codes in our dataset. Use the stations and airports below for the analysis, and create a new column called `airport` that maps the weather station to the airport name.

````python
weather = pd.read_csv("daily-summaries.csv")

station = ['USW00023174','USW00012960','USW00003017','USW00094846',
           'USW00013874','USW00023234','USW00003927','USW00023183','USW00013881'] 
airports = ['LAX', 'IAH', 'DEN', 'ORD', 'ATL', 'SFO', 'DFW', 'PHX', 'CLT']

### Map weather stations to airport code

station_map =  dict(zip(station, airports)) 

# Mapping the dictionary station_map to the data frame.

weather['airport'] =  weather['STATION'].map(station_map)
````

````python
# To view the new dataframe, weather
weather
````

Create another column called `MONTH` from the `DATE` column.

````python
weather['MONTH'] = weather['DATE'].apply(lambda x: x.split('-')[1])
weather.head()
````

Analyze and handle the `SNOW` and `SNWD` columns for missing values using `fillna()`. Use the `isna()` function to check the missing values for all the columns.

````python
weather.isna().sum()
````

There are 5478 missing values in the columns 'SNOW' and 'SNWD' out of 16434 rows. So we can replace these na values using mean of the columns. 

````python
weather.SNOW.fillna(weather['SNOW'].mean(), inplace=True)
weather.SNWD.fillna(weather['SNWD'].mean(), inplace=True)
weather.isna().sum()
````

**Question**: Print the index of the rows that have missing values for TAVG, TMAX, TMIN.

We use the `isna()` function to find the rows that are missing, and then use the list on the idx variable to get the index.

There are 62 missing rows in the variable "TAVG" followed by 20 missing rows in each of the columns "TMAX" and "TMIN". 

````python
idx = np.array([i for i in range(len(weather))])
TAVG_idx = idx[weather['TAVG'].isna()]
TMAX_idx = idx[weather['TMAX'].isna()]
TMIN_idx = idx[weather['TMIN'].isna()]
TAVG_idx
````

We can replace the missing TAVG, TMAX, and TMIN with the average value for a particular station/airport. Because the consecutive rows of TAVG_idx are missing, replacing with a previous value would not be possible. Instead, replace it with the mean. 
We use the `groupby` function to aggregate the variables with a mean value.

````python
weather_impute = weather.groupby(['STATION','MONTH']).agg({'TAVG':'mean','TMAX':'mean', 'TMIN':'mean' }).reset_index()
weather_impute.head()
````

````python
weather
````

Merge the mean data with the weather data.

````python
### get the yesterday's data
weather = pd.merge(weather, weather_impute,  how='left', left_on=['MONTH','STATION'], right_on = ['MONTH','STATION'])\
.rename(columns = {'TAVG_y':'TAVG_AVG',
                   'TMAX_y':'TMAX_AVG', 
                   'TMIN_y':'TMIN_AVG',
                   'TAVG_x':'TAVG',
                   'TMAX_x':'TMAX', 
                   'TMIN_x':'TMIN'})
````

Check for missing values again.

````python
weather.TAVG[TAVG_idx] = weather.TAVG_AVG[TAVG_idx]
weather.TMAX[TMAX_idx] = weather.TMAX_AVG[TMAX_idx]
weather.TMIN[TMIN_idx] = weather.TMIN_AVG[TMIN_idx]
weather.isna().sum()
````

We can see that there are no missing values in the weather data now.
